```python
cols = [
    'date_mutation',
    'valeur_fonciere',
    'surface_reelle_bati',
    'code_commune',
    'type_local',
    'nombre_pieces_principales',
    'longitude',
    'latitude',
    'code_postal',
    'id_mutation'
]

def prepare_and_agg(df):

    os.makedirs('raw', exist_ok=True)
    
    df = df[cols].copy()
    df['prix_m2'] = df['valeur_fonciere'] / df['surface_reelle_bati']
    df['annee'] = pd.to_datetime(df['date_mutation']).dt.year
    
    df_agg = df.groupby(['code_commune', 'annee']).agg({
        'prix_m2': 'mean',
        'surface_reelle_bati': 'mean',
        'id_mutation': 'count'
    }).rename(columns={'id_mutation':'nb_ventes'}).reset_index()
    
    annee= df['annee'].unique()[0] 
    output_path = os.path.join('raw', f'csv_agg_{annee}.csv')
    
    df_agg.to_csv(output_path, index=False, sep=';', encoding='utf-8')
    
    print(f"Fichier CSV enregistré ici : {output_path}")
    return df_agg
```

```python
vf_annee = df_merged.groupby(['annee', 'code_postal'], dropna=False).agg(
    prix_m2=('prix_m2', 'mean'),
    dep_code=('dep_code', 'first'),
    dep_nom=('dep_nom', 'first'),
    reg_code=('reg_code', 'first'),
    reg_nom=('reg_nom', 'first'),
    population=('population', 'sum'),
    superficie_km2=('superficie_km2', 'sum'),  
    zone_emploi=('zone_emploi', 'first'),
    taux_global_tfb=('taux_global_tfb', 'mean'),
    taux_global_tfnb=('taux_global_tfnb', 'mean'),
    taux_plein_teom=('taux_plein_teom', 'mean'),
    taux_global_th=('taux_global_th', 'mean'),
    nb_ventes=('nb_ventes', 'sum')
).reset_index()


vf_annee = vf_annee.sort_values(['code_postal', 'annee'])
vf_annee['variation'] = vf_annee.groupby('code_postal')['prix_m2'].pct_change()
vf_annee = vf_annee[vf_annee['annee'] != 2020].reset_index(drop=True)


"""
def market_trend(x, seuil=0.02):  
    if x > seuil:
        return 'hausse'
    elif x < -seuil:
        return 'baisse'
    else:
        return 'stable'
    
vf_annee['y'] = vf_annee['variation'].apply(market_trend)
"""
```

```python
df_inflation = pd.read_csv(taux_inflation, sep=';', parse_dates=['Date'])
print(df_inflation.columns.tolist())
df_inflation.rename(columns={"Taux d’inflation": "taux_inflation"}, inplace=True)
```

```python
df_inflation['annee'] = df_inflation['Date'].dt.year
df_inflation['taux_inflation'] = df_inflation["taux_inflation"] / 100
```

```python
df_inflation['taux_inflation']
```

```python
inflation_par_annee = (
    df_inflation
    .sort_values('Date')
    .groupby('annee')['taux_inflation']
    .last()        
    .to_dict()
)

if 2025 not in inflation_par_annee:
    inflation_par_annee[2025] = inflation_par_annee[2024]
    
vf_annee['taux_inflation'] = vf_annee['annee'].map(inflation_par_annee)

def market_trend_dynamic(row):
    x = row['variation']       
    seuil = row['taux_inflation']  

    if pd.isna(x) or pd.isna(seuil):
        return None

    if x > seuil:
        # Le prix augmente AU-DELÀ de l'inflation → hausse réelle
        return 'hausse'
    elif x < - seuil :
        # Le prix baisse en valeur nominale → baisse réelle (peu importe l'inflation)
        return 'baisse'
    else:
        # Le prix augmente mais moins que l'inflation → stable en termes réels
        return 'stable'

vf_annee['y'] = vf_annee.apply(market_trend_dynamic, axis=1)
```
